# LabelEncoder

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['title'] = label_encoder.fit_transform(df['title'])
df['uploaded_by'] = label_encoder.fit_transform(df['uploaded_by'])
df['last_updated'] = label_encoder.fit_transform(df['last_updated'])
df['usability'] = label_encoder.fit_transform(df['usability'])
df['files'] = label_encoder.fit_transform(df['files'])
df['size'] = label_encoder.fit_transform(df['size'])
df['upvotes'] = label_encoder.fit_transform(df['upvotes'])
df['badge'] = label_encoder.fit_transform(df['badge'])
```

# SEPARATING DATASET

```python
df1 = df.iloc[0:340]
```

```python
df1
```

```python
df2 = df.iloc[341:671]
```

```python
df2
```

```python
df3 = df.iloc[671:1000]
```

```python
df3
```

```python
x = df1.iloc[:, :-1]
y = df1.iloc[:, -1]
```

```python
x
```

```python
y
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 10)
```

```python
y_test
```

# DECISION TREES

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=1234)
dtc.fit(x_train, y_train)
```

```python
y_predict = dtc.predict(x_test)
```

```python
y_predict
```

# confusion_matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
score = dtc.score(x_test, y_test)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score
accuracy = accuracy_score(y_test, y_predict)
print('Accuracy',accuracy)
recall = recall_score(y_test,y_predict,average = 'macro')
print('Recall',recall)
f1score = f1_score(y_test, y_predict,average = 'macro')
print('f1 score',f1score)
```
